```python
from PIL import Image
import json 

def resize_with_bounding_box(input_path, json_file, new_size, output_path):
    try:
        # Open the image file
        with Image.open(input_path) as img:
            # Resize the image
            resized_img = img.resize(new_size)
        with open(json_file, 'r') as file:
            # Load the JSON data into a dictionary
            data_dict = json.load(file)
        print(data_dict["form"])
        for i in range(len(data_dict['form'])):
            print(i)
            x1,y1,x2,y2 = data_dict['form'][i]['box']
            print(x1,y1,x2,y2)
            
            # # Scale the bounding box coordinates
            # x1, y1, x2, y2 = bounding_box
            x1_scaled = int(x1 * (new_size[0] / img.width))
            y1_scaled = int(y1 * (new_size[1] / img.height))
            x2_scaled = int(x2 * (new_size[0] / img.width))
            y2_scaled = int(y2 * (new_size[1] / img.height))
            data_dict['form'][i]['box'] = [x1_scaled,y1_scaled,x2_scaled,y2_scaled]
            for j in range(len(data_dict['form'][i]['words'])):
                x1,y1,x2,y2 = data_dict['form'][i]['words'][j]["box"]
                x1_scaled = int(x1 * (new_size[0] / img.width))
                y1_scaled = int(y1 * (new_size[1] / img.height))
                x2_scaled = int(x2 * (new_size[0] / img.width))
                y2_scaled = int(y2 * (new_size[1] / img.height))
                data_dict['form'][i]['words'][j]["box"] = [x1_scaled,y1_scaled,x2_scaled,y2_scaled]
                        
                
            print(data_dict['form'][i]['box'])
            print(data_dict['form'][i]['words'])
            # # Save the resized image
        image_output_path = output_path+"images/"+input_path.split("/")[-1]
        resized_img.save(image_output_path)
        # Specify the path to save the JSON file
        
        json_file_path = output_path+"annotations/"+ json_file.split("/")[-1]

        # Open the JSON file in write mode
        with open(json_file_path, 'w') as file:
            # Dump the dictionary into the file
            json.dump(data_dict, file)
        # print(f"Image resized and saved to {output_path}")
            # break
            # print(f"Scaled Bounding Box: ({x1_scaled}, {y1_scaled}, {x2_scaled}, {y2_scaled})")
    except Exception as e:
        print(f"Error: {e}")
```
